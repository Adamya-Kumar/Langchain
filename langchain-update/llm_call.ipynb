```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

temperature 0 mean not creativite and 1 mean very cerative response.

```python
llm=ChatGoogleGenerativeAI(model="gemini-3.5-flash",temperature=0.7)

response=llm.invoke("what is linear regression in ml?")
```

##### The reason you see both .text and .content is that .text is a convenience shortcut that gives you just the clean string response, while .content provides the raw, complex data structure containing everything the model generated.

```python
print(response.text)
```

```python
print(response.content)
```

("role", "content") (Tuple): The official, safe standard used in documentation. It cannot be accidentally changed by your code.["role", "content"] (List): The flexible option. Use this when loading messages dynamically from JSON files or databases.

```python
response=llm.invoke([
    ["system","Your are helpfull assitent?tone polite and simple"],
    ["human","who many moon pluto planet?"]
])
```

```python
print(response.text)
```

## GROQ LLM-CALL
 

```python
from langchain_groq import ChatGroq
```

```python
groq_llm=ChatGroq(model="qwen/qwen3-32b",reasoning_format="parsed")

messages=[
    ("system","You are helpfull assistance?"),
    ("human","who many moon pluto planet?")
]

ai_message=groq_llm.invoke(messages)

print(ai_message.text)
```
